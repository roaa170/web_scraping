```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = 'https://books.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text , 'html.parser')
soup
```

```python
with open('books.csv', 'w' ,encoding='utf-8' ,newline="") as file:
  writer = csv.writer(file)
  writer.writerow(['title' , 'price' , 'rating' , 'availability'])
  books = soup.find_all('article' , attrs={'class':'product_pod'})
 # books
  for book in books:
    title = book.h3.a['title']
    price = book.find('p', class_='price_color').text.strip()
    rating = book.p['class'][1]
    availability = book.find('p', class_='instock availability').text.strip()
    writer.writerow([title , price , rating , availability])
    print(f'📘 {title}\n⭐ {rating} \n💲 {price} \n📦 {availability} \n{"*"*50} ')
```
